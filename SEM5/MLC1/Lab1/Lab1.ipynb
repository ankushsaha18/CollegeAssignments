```python
import numpy as np
arr = np.arange(1,21)

def isPrime(n):
    if(n < 2):
        return False
    for i in range(2,int(np.sqrt(n))+1):
        if n%i == 0:
            return False
    return True

primes = [x for x in arr if isPrime(x)]
mean = np.mean(primes)
variance = np.var(primes)
print(primes)
print(mean)
print(variance)
```

```python
import pandas as pd

data = {
    'Name': ['Amit', 'Priya', 'Rahul', 'Sneha', 'Karan'],
    'Math': [85, 92, 78, 88, 95],
    'Science': [90, 89, 84, 91, 88],
    'English': [88, 94, 80, 86, 90]
}

df = pd.DataFrame(data)

df['Total'] = df['Math'] + df['Science'] + df['English']
df['Average'] = df['Total']/3

# Find topper without idxmax()
top_avg = df['Average'].max()
topper = df[df['Average'] == top_avg]

print(df)
print("Topper:", topper['Name'].values[0])
print("Average Marks:", topper['Average'].values[0])
```

```python
import numpy as np

# (a) Simulate 1000 coin tosses (1 = Head, 0 = Tail)
tosses = np.random.randint(0, 2, 1000)

# (b) Count frequency of heads and tails
heads = np.sum(tosses == 1)
tails = np.sum(tosses == 0)

# (c) Estimate probability of heads
p_heads = heads / 1000

print("Heads:", heads)
print("Tails:", tails)
print("Estimated Probability of Heads:", p_heads)
```

```python
import numpy as np

# (a) Create a 3×3 array with values 1 to 9
A = np.arange(1, 10).reshape(3, 3)
print("A =\n", A)

# (b) Transpose and inverse
A_T = A.T
print("\nTranspose A^T =\n", A_T)

A_inv = np.linalg.inv(A)
print("\nInverse A^-1 =\n", A_inv)

# (c) Verify A × A⁻¹ ≈ I
I = np.dot(A,A_inv)
print("\nA × A^-1 =\n", I)

print("Close? = ",np.allclose(I,np.eye(3)))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(43)

df = pd.DataFrame({
    "Math": np.random.randint(40, 100, 100),
    "Science": np.random.randint(35, 100, 100),
    "English": np.random.randint(30, 100, 100)
})

# (a)
df["Average"] = df[["Math", "Science", "English"]].mean(axis=1)

top_5 = df.sort_values("Average", ascending=False).head(5)
print("\nTop 5 Performers:")
print(top_5)

# (b)
subject_avg = df[["Math", "Science", "English"]].mean()
subject_std = df[["Math", "Science", "English"]].std()

print("\nSubject-wise Averages:")
print(subject_avg)

print("\nSubject-wise Standard Deviations:")
print(subject_std)

print("\nHighest variation in:", subject_std.idxmax())

# (c)
plt.boxplot(df)
plt.title("Boxplot of Student Marks")
plt.ylabel("Marks")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.DataFrame({
    "Product": ["A", "B", "C", "A", "B", "D", "C", "A"],
    "Quantity": [10, 5, 8, 12, 7, 3, 9, 4],
    "Price":    [100, 200, 150, 100, 200, 50, 150, 100],
    "Region":   ["North", "South", "East", "West", "North", "South", "East", "West"]
})


# (a)
df["Total"] = df["Quantity"] * df["Price"]
overall_revenue = df["Total"].sum()
print("\nData with Total Column:")
print(df)

print("\nOverall Revenue:", overall_revenue)


# (b)
region_sales = df.groupby("Region")["Total"].sum()
print("\nRegion-wise Total Sales:")
print(region_sales)
top_region = region_sales.idxmax()
print("\nRegion contributing the most sales:", top_region)


# (c)
product_sales = df.groupby("Product")["Total"].sum()
print("\nTotal Sales per Product:")
print(product_sales)

plt.bar(product_sales.index,product_sales.values)
plt.title("Histogram Plot of Total Sales per Product")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.show()

lowest_sales_product = product_sales.idxmin()
print("\nUnderperforming Product (lowest revenue):", lowest_sales_product)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("auto_mpg.csv")


# (a)
numeric_df = df.drop(columns=["car_name"])
corr_with_mpg = numeric_df.corrwith(df["mpg"]).sort_values(ascending=False)

print("\nCorrelation of each feature with MPG:")
print(corr_with_mpg)

print("\nStrongest positive influence:", corr_with_mpg.idxmax(), corr_with_mpg.max())
print("Strongest negative influence:", corr_with_mpg.idxmin(), corr_with_mpg.min())


# (b)
plt.scatter(df["weight"], df["mpg"])
plt.title("Scatter Plot: Weight vs MPG")
plt.xlabel("Vehicle Weight")
plt.ylabel("Miles Per Gallon (MPG)")
plt.show()
```
