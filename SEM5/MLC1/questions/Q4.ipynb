```python
from sklearn.datasets import load_diabetes
import pandas as pd

data = load_diabetes()
X, y = data.data, data.target

df = pd.DataFrame(X, columns=data.feature_names)
df['target'] = y
print("Missing values:\n",df.isnull().sum())
df.dropna(inplace=True)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)

from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Predictions:", y_pred)
```

```python
import numpy as np
matrix = np.random.rand(5,4)

from sklearn.decomposition import TruncatedSVD
svd = TruncatedSVD(n_components=2)
reduced_matrix = svd.fit_transform(matrix)
print("Reduced matrix:\n", reduced_matrix)

matrix_T = svd.transform(matrix)
print(matrix_T.shape)
print(reduced_matrix.shape)
print(matrix.shape)
```

```python
from scipy import stats

p = 0.3
n = 2000
x = np.arange(0, n+1)

pmf = stats.binom.pmf(x, n, p)

print("Mean:", stats.binom.mean(n, p))
print("Variance:", stats.binom.var(n, p))

cdf = stats.binom.cdf(x,n,p)
import matplotlib.pyplot as plt
plt.plot(x, cdf)
plt.show()
```

```python
x = np.array([15,23,18,23,24,22,22,19,19,16,24,11,24,16,23]).reshape(-1,1)
y=np.array([49,63,58,60,58,61,60,63,60,52,62,30,59,49,68])

from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x,y)
y_pred = reg.predict(x)
print("Predicted values:", y_pred)

print("Slope:", reg.coef_[0])   
print("Intercept:", reg.intercept_)

from sklearn.metrics import r2_score
r2 = r2_score(y, y_pred)
print("R-squared:", r2)

import matplotlib.pyplot as plt
plt.scatter(x, y, color='blue', label='Actual data')
plt.plot(x, y_pred, color='red', label='Regression line')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend()
plt.show()
```
