```python
def generateMatrix(key):
    # Convert Upper casev& replace j with i
    key = key.upper().replace("J", "I")
    matrix = []
    # Remove Duplicate Letter
    used = set()
    
    for char in key:
        if char not in used:
            # Append Alphate Letter
            matrix.append(char)
            used.add(char)

    for char in "ABCDEFGHIKLMNOPQRSTUVWXYZ":  
        if char not in used:
            # Append the Remaining Character
            matrix.append(char)

    # The 25 character string is placed row wise in a 5X5 matrix
    return [matrix[i:i+5] for i in range(0, 25, 5)]

def textPairing(text):
    # Convert Upper case & replace j with i
    text = text.upper().replace("J", "I")

    i = 0
    result = ""
    # split into pair
    while i < len(text):
        a = text[i]
        b = text[i+1] if i+1 < len(text) else 'X'

        if a == b:
            #add X for reapeated letters
            result += a + 'X'
            i += 1
        else:
            result += a + b
            i += 2
            

    #add X for odd length letters
    if len(result) % 2 != 0:
        result += 'X'

    return result

# Find the Position of Letter
def position(matrix, char):
    for i in range(5):
        for j in range(5):
            if matrix[i][j] == char:
                return i, j


def encryptPlayfair(pt, key):
    matrix = generateMatrix(key)
    print("\nKey Matrix")
    for i in matrix:
        print(i)
    text = textPairing(pt)
    print("\nDigraphs--")
    for i in range(0, len(text)-1, 2):
        print(text[i], text[i+1])
    ct = ""

    for i in range(0, len(text)-1, 2):
        a, b = text[i], text[i+1]
        r1, c1 = position(matrix, a)
        r2, c2 = position(matrix, b)

        #Same Row : replace with its Right
        if r1 == r2:  
            ct += matrix[r1][(c1 + 1) % 5]
            ct += matrix[r2][(c2 + 1) % 5]

        #Same Column : replace with its Below
        elif c1 == c2:  
            ct += matrix[(r1 + 1) % 5][c1]
            ct += matrix[(r2 + 1) % 5][c2]

        #Rectangle : replace each letter in same row but column of the other letter
        else:  
            ct += matrix[r1][c2]
            ct += matrix[r2][c1]

    return ct


def decryptPlayfair(ct, key):
    matrix = generateMatrix(key)
    print("\nKey Matrix")
    for i in matrix:
        print(i)
    text = ct.upper()
    print("\nDigraphs--")
    for i in range(0, len(text)-1, 2):
        print(text[i], text[i+1])
    pt = ""

    for i in range(0, len(text)-1, 2):
        a, b = text[i], text[i+1]
        r1, c1 = position(matrix, a)
        r2, c2 = position(matrix, b)

        #Same Row : replace with its Right
        if r1 == r2: 
            pt += matrix[r1][(c1 - 1) % 5]
            pt += matrix[r2][(c2 - 1) % 5]

        #Same Column : replace with its Below
        elif c1 == c2: 
            pt += matrix[(r1 - 1) % 5][c1]
            pt += matrix[(r2 - 1) % 5][c2]

        #Rectangle : replace each letter in same row but column of the other letter
        else:  
            pt += matrix[r1][c2]
            pt += matrix[r2][c1]

    return pt

pt = input("Enter plain text: ")
key = input("Enter key: ")
print("\nCipher Text: ",encryptPlayfair(pt, key))

ct = input("\n\nEnter cipher text: ")
key = input("Enter key: ")
print("\nPlain Text: ",decryptPlayfair(ct, key))
```

```python
def encryptVigenere(pt, key):
    # Convert plaintext to uppercase
    pt = pt.upper()  
    # Convert key to uppercase 
    key = key.upper() 
    
    nk = ""  
    i = 0

    # Generate repeating key equal to length of plaintext
    for _ in range(len(pt)):
        nk += key[i]
        i = (i + 1) % len(key)  
    ct = "" 

    for j in range(len(pt)):
        # Convert plaintext and key letters to number (0–25)
        p = ord(pt[j]) - ord('A')   
        k = ord(nk[j]) - ord('A')   

        # Apply Vigenere formula: (P + K) mod 26
        letter = chr((p + k) % 26 + ord('A'))

        ct += letter  

    return ct


def decryptVigenere(ct, key):
    # Convert ciphertext to uppercase
    ct = ct.upper()   
    # Convert key to uppercase
    key = key.upper() 

    nk = ""  
    i = 0

    # Generate repeating key
    for _ in range(len(ct)):
        nk += key[i]
        i = (i + 1) % len(key)

    pt = ""  

    for j in range(len(ct)):
        # Convert ciphertext and key letters to number
        c = ord(ct[j]) - ord('A')   
        k = ord(nk[j]) - ord('A')   

        # Apply Vigenere formula: (C - K) mod 26
        letter = chr((c - k + 26) % 26 + ord('A'))

        pt += letter  

    return pt


pt = input("Enter plain text: ")
key = input("Enter key: ")
print("\nCipher Text:", encryptVigenere(pt, key))

ct = input("\n\nEnter cipher text: ")
key = input("Enter key: ")
print("\nPlain Text:", decryptVigenere(ct, key))
```
