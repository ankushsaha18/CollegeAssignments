```python
# 2. Create a custom Python class for managing a bank account with basic functionalities like deposit and withdrawal?

class BankAccount:

    def __init__(self, name, balance=0):
        self.name = name
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print("Deposited:", amount)

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance")
        else:
            self.balance -= amount
            print("Withdrawn:", amount)

    def display(self):
        print("Account Holder:", self.name)
        print("Balance:", self.balance)


acc = BankAccount("Ankush", 1000)
acc.deposit(500)
acc.withdraw(200)
acc.display()

''' Ankush Saha , 2341019012 , 51 '''
```

```python
# 3. Create a Book class that contains multiple Chapters, where each Chapter has a title and page count.
# Write code to initialize a Book object with three chapters and display the total page count of the book.

class Chapter:

    def __init__(self, title, pages):
        self.title = title
        self.pages = pages


class Book:

    def __init__(self, name, chapters):
        self.name = name
        self.chapters = chapters

    def total_pages(self):
        total = 0
        for ch in self.chapters:
            total += ch.pages
        return total


c1 = Chapter("Intro", 20)
c2 = Chapter("OOP", 30)
c3 = Chapter("Advanced Python", 40)

book = Book("Python Programming", [c1, c2, c3])

print("Total Pages:", book.total_pages())

''' Ankush Saha , 2341019012 , 51 '''
```

```python
# 5. Write a Python program using a Time class to input a given time in 24-hour format and convert it
# to a 12-hour format with AM/PM.

class Time:

    def __init__(self, time_str):
        self.time_str = time_str

    def is_valid(self):
        try:
            h, m, s = map(int, self.time_str.split(":"))
            return 0 <= h < 24 and 0 <= m < 60 and 0 <= s < 60
        except:
            return False

    def convert(self):

        if not self.is_valid():
            return "Invalid Time Format"

        h, m, s = map(int, self.time_str.split(":"))

        period = "AM"
        if h >= 12:
            period = "PM"

        h = h % 12
        if h == 0:
            h = 12

        return f"{h:02}:{m:02}:{s:02} {period}"


t = Time("18:30:20")
print(t.convert())

''' Ankush Saha , 2341019012 , 51 '''
```

```python
# 7. Write a Python program to simulate a card game using object-oriented principles.

import random

class Card:
    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank

    def __str__(self):
        return f"{self.rank} of {self.suit}"


class Deck:
    def __init__(self):
        suits = ["Hearts","Diamonds","Clubs","Spades"]
        ranks = ["A","2","3","4","5"]
        self.cards = [Card(s,r) for s in suits for r in ranks]
        
    def shuffle(self):
        random.shuffle(self.cards)
    def deal(self, players, num):
        for _ in range(num):
            for p in players:
                p.hand.append(self.cards.pop())


class Player:
    def __init__(self, name):
        self.name = name
        self.hand = []

    def show(self):
        print(self.name, [str(c) for c in self.hand])


deck = Deck()
deck.shuffle()

p1 = Player("Player1")
p2 = Player("Player2")

deck.deal([p1,p2],3)

p1.show()
p2.show()

''' Ankush Saha , 2341019012 , 51 '''
```

```python
# 8. Write a Python program that defines a base class Vehicle and subclass Car demonstrating inheritance.

class Vehicle:

    def __init__(self, make, model):
        self.make = make
        self.model = model

    def display_info(self):
        print(self.make, self.model)


class Car(Vehicle):

    def __init__(self, make, model, num_doors):
        super().__init__(make, model)
        self.num_doors = num_doors

    def display_info(self):
        super().display_info()
        print("Doors:", self.num_doors)


v = Vehicle("Honda","Bike")
c = Car("Toyota","Corolla",4)

v.display_info()
c.display_info()

''' Ankush Saha , 2341019012 , 51 '''
```

```python
# 9. Write a Python program demonstrating polymorphism by creating a base class Shape with a method
# area(), and two subclasses Circle and Rectangle that override the area() method.

import math

class Shape:
    def area(self):
        pass


class Circle(Shape):

    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius * self.radius


class Rectangle(Shape):

    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width


shapes = [Circle(5), Rectangle(4, 6)]

for s in shapes:
    print("Area:", s.area())

''' Ankush Saha , 2341019012 , 51 '''
```

```python
# 10. Implement the CommissionEmployee class with init, earnings, and repr methods.
# Include properties for personal details and sales data.

class CommissionEmployee:
    def __init__(self, name, sales, commission_rate):
        if sales < 0 or commission_rate < 0:
            raise ValueError("Negative values are not allowed")
        self.name = name
        self.sales = sales
        self.commission_rate = commission_rate

    def earnings(self):
        return self.sales * self.commission_rate

    def __repr__(self):
        return f"Employee: {self.name}, Sales: {self.sales}, Rate: {self.commission_rate}"


emp = CommissionEmployee("Ankush", 10000, 0.1)

print(emp)
print("Earnings:", emp.earnings())

emp.sales = 20000

print("Updated Earnings:", emp.earnings())

''' Ankush Saha , 2341019012 , 51 '''
```

```python
# 11. What is duck typing in Python? Write a Python program demonstrating duck typing by creating a
# function describe() that accepts any object with a speak() method.


'''

Duck typing is a concept in Python where the type of an object is not checked explicitly.
Instead, the focus is on whether the object has the required methods or behavior.

The name comes from the phrase:
"If it walks like a duck and quacks like a duck, then it is a duck."

In duck typing, an object is considered suitable if it has the required method or property,
regardless of its actual class.

Explanation of the program:
In the program, the function describe() calls the speak() method on the object passed to it.
Both Dog and Robot classes implement a speak() method. When objects of these classes are
passed to the describe() function, the function works correctly without checking whether
the object is of type Dog or Robot. Python only checks if the object has the speak() method.

'''

class Dog:
    def speak(self):
        print("Dog barks")

class Robot:
    def speak(self):
        print("Robot speaking")

def describe(obj):
    obj.speak()

d = Dog()
r = Robot()
describe(d)
describe(r)

''' Ankush Saha , 2341019012 , 51 '''
```

```python
# 12. WAP to overload the + operator to perform addition of two complex numbers using a custom Complex class.

class Complex:
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag

    def __add__(self, other):
        return Complex(self.real + other.real, self.imag + other.imag)

    def display(self):
        print(self.real, "+", self.imag, "i")


c1 = Complex(3, 4)
c2 = Complex(2, 5)

c3 = c1 + c2
c3.display()

''' Ankush Saha , 2341019012 , 51 '''
```

```python
# 17. Scope Resolution: object’s namespace → class namespace → global namespace → built-in namespace.

species = "Global Species"

class Animal:
    species = "Class Species"
    def __init__(self, species):
        self.species = species

    def display_species(self):
        print("Instance species:", self.species)
        print("Class species:", Animal.species)
        print("Global species:", globals()['species'])

a = Animal("Instance Species")
a.display_species()


'''

Explanation:

Python follows a scope resolution rule similar to the LEGB rule
(Local → Enclosing → Global → Built-in). In object-oriented programs,
attribute lookup happens in the following order:

1. Object (Instance) Namespace
Python first checks the attributes of the object itself.
Here, self.species refers to the instance variable created in the constructor,
which is "Instance Species".

2. Class Namespace
If the attribute is not found in the object, Python checks the class attributes.
Animal.species refers to the class variable defined in the class, which is
"Class Species".

3. Global Namespace
If the attribute is still not found, Python checks the global variables defined
outside the class. Here, globals()['species'] refers to the global variable
"Global Species".

4. Built-in Namespace
Finally, Python checks built-in names such as len(), print(), etc.

Thus, the program prints the instance value first, then the class value,
and finally the global value.

'''

''' Ankush Saha , 2341019012 , 51 '''
```

```python
# 18. Write a Python program using a lambda function to convert temperatures from Celsius to Kelvin,
# store the data in a tabular format using pandas, and visualize the data using a plot.

import pandas as pd
import matplotlib.pyplot as plt

celsius = [0, 10, 20, 30, 40]
kelvin = list(map(lambda x: x + 273.15, celsius))

df = pd.DataFrame({
    "Celsius": celsius,
    "Kelvin": kelvin
})

print(df)

plt.plot(df["Celsius"], df["Kelvin"])
plt.xlabel("Celsius")
plt.ylabel("Kelvin")
plt.title("Temperature Conversion")

plt.show()

''' Ankush Saha , 2341019012 , 51 '''
```
