```python
# 1. Write a recursive function power(base, exponent) that, when called,
# returns base^exponent. Example: power(3,4) = 3*3*3*3.
# Both base and exponent are user-defined and exponent >= 1.

def power(base, exponent):
    if exponent == 1:
        return base
    return base * power(base, exponent - 1)

print(power(3, 4))

''' Ankush Saha, 2341019012 ,51 '''
```

```python
# 2. The greatest common divisor of integers x and y is the largest integer
# that evenly divides into both x and y. Write and test a recursive function gcd.
# If y == 0, gcd(x, y) = x
# otherwise gcd(x, y) = gcd(y, x%y)

def gcd(x, y):
    if y == 0:
        return x
    return gcd(y, x % y)

print(gcd(48, 18))

''' Ankush Saha, 2341019012 ,51 '''
```

```python
# 4. Implement a recursive solution for computing the nth Fibonacci number.
# Analyze its time complexity. Propose a more efficient solution and compare.

def fibonacci_recursive(n):
    if n <= 1:
        return n
    return fibonacci_recursive(n-1) + fibonacci_recursive(n-2)

def fibonacci_iterative(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a

print(fibonacci_recursive(6))
print(fibonacci_iterative(6))

# Recursive Time Complexity = O(2^n)
# Iterative Time Complexity = O(n)

''' Ankush Saha, 2341019012 ,51 '''
```

```python
# 5. Given an array of N elements, not necessarily in ascending order,
# devise an algorithm to find the kth largest one.
# It should run in O(N) time on random inputs.

def partition(arr, left, right):
    pivot = arr[right]
    i = left
    for j in range(left, right):
        if arr[j] >= pivot:
            arr[i], arr[j] = arr[j], arr[i]
            i += 1
    arr[i], arr[right] = arr[right], arr[i]
    return i

def quickselect(arr, left, right, k):
    if left <= right:
        pivot_index = partition(arr, left, right)
        if pivot_index == k:
            return arr[pivot_index]
        elif pivot_index > k:
            return quickselect(arr, left, pivot_index - 1, k)
        else:
            return quickselect(arr, pivot_index + 1, right, k)


arr = [7, 10, 4, 3, 20, 15]
k = 3
print(quickselect(arr, 0, len(arr)-1, k-1))

''' Ankush Saha, 2341019012 ,51 '''
```

```python
# 7. Given N points on a circle, centered at the origin, design an algorithm
# that determines whether there are two points that are antipodal,
# i.e., the line connecting the two points goes through the origin.
# Algorithm should run in time proportional to NlogN.

def has_antipodal(points):
    points_set = set(points)
    for x, y in points:
        if (-x, -y) in points_set:
            return True
    return False

points = [(1, 2), (-1, -2), (3, 4)]
print(has_antipodal(points))

''' Ankush Saha, 2341019012 ,51 '''
```

```python
# 8. Write a Python function quick sort that implements the quicksort algorithm.
# The function should include a helper function quick_sort_helper
# to handle recursion and sort array in-place.

def quick_sort(arr):
    def partition(arr, low, high):
        pivot = arr[low]
        left = low + 1
        right = high
        while True:
            while left <= right and arr[left] <= pivot:
                left += 1
            while left <= right and arr[right] >= pivot:
                right -= 1
            if left > right:
                break
            arr[left], arr[right] = arr[right], arr[left]
        arr[low], arr[right] = arr[right], arr[low]
        return right

    def quick_sort_helper(arr, low, high):
        if low < high:
            pivot_index = partition(arr, low, high)
            quick_sort_helper(arr, low, pivot_index-1)
            quick_sort_helper(arr, pivot_index+1, high)

    quick_sort_helper(arr, 0, len(arr)-1)

arr = [37, 2, 6, 4, 89, 8, 10, 12, 68, 45]
quick_sort(arr)
print(arr)

''' Ankush Saha, 2341019012 ,51 '''
```

```python
# 9. Sort the following personalities based on net worth using
# Selection Sort, Bubble Sort and Insertion Sort and present result as dictionary.

people = [
    ("Elon Musk", 433.9),
    ("Jeff Bezos", 239.4),
    ("Mark Zuckerberg", 211.8),
    ("Larry Ellison", 204.6),
    ("Bernard Arnault & Family", 181.3),
    ("Larry Page", 161.4)
]

def selection_sort(data):
    arr = data[:]
    for i in range(len(arr)):
        min_index = i
        for j in range(i+1, len(arr)):
            if arr[j][1] < arr[min_index][1]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return dict(arr)

print(selection_sort(people))

''' Ankush Saha, 2341019012 ,51 '''
```

```python
# 10. Use Merge Sort to sort a list of strings alphabetically.

def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr)//2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

arr = ['apple', 'orange', 'banana', 'grape']
print(merge_sort(arr))

''' Ankush Saha, 2341019012 ,51 '''
```

```python
# 11. Without using sorted(), merge two pre-sorted lists into
# a single sorted list using Merge Sort logic.

def merge_lists(list1, list2):
    i = j = 0
    merged = []
    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            merged.append(list1[i])
            i += 1
        else:
            merged.append(list2[j])
            j += 1
    merged.extend(list1[i:])
    merged.extend(list2[j:])
    return merged

print(merge_lists([1,3,5,7], [2,4,6,8]))

''' Ankush Saha, 2341019012 ,51 '''
```
