```python
# 5. Write a Python script to perform NLP tasks (fixed version)

import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords

# Required downloads(once)
# nltk.download('punkt')
# nltk.download('stopwords')
# nltk.download('wordnet')
# nltk.download('omw-1.4')

text = """Natural Language Processing enables machines to understand and process human languages.
It is a fascinating field with numerous applications, such as chatbots and language translation."""

# Tokenization
words = word_tokenize(text)
sentences = sent_tokenize(text)

# Stemming
ps = PorterStemmer()
stemmed = [ps.stem(word) for word in words]

# Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized = [lemmatizer.lemmatize(word) for word in words]

# Stopword removal
stop_words = set(stopwords.words('english'))
filtered = [word for word in words if word.lower() not in stop_words]

print("Words:", words)
print("Sentences:", sentences)
print("Stemmed:", stemmed)
print("Lemmatized:", lemmatized)
print("Filtered:", filtered)

''' Ankush Saha , 2341019012 , 51 '''
```

```python
# 9. (Sentiment of a News Article with the NaiveBayesAnalyzer) 
# Repeat the previous exercise but use the NaiveBayesAnalyzer for sentiment analysis.

from textblob import TextBlob
from textblob.sentiments import NaiveBayesAnalyzer
import requests
from bs4 import BeautifulSoup

url = "https://www.bbc.com/news/technology-65000000"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
article_text = ' '.join(p.get_text() for p in soup.find_all('p'))

blob = TextBlob(article_text, analyzer=NaiveBayesAnalyzer())

print("Overall Sentiment:", blob.sentiment)

for sentence in blob.sentences:
    sent_blob = TextBlob(str(sentence), analyzer=NaiveBayesAnalyzer())
    print(sentence, "->", sent_blob.sentiment)

''' Ankush Saha , 2341019012 , 51 '''
```

```python
# 12. Write a Python program that takes a word as input and returns:
# • Its definition
# • Its synonyms
# • Its antonyms(if available)

from nltk.corpus import wordnet

word = input("Enter word: ")
synsets = wordnet.synsets(word)
if synsets:
    print("Definition:", synsets[0].definition())
synonyms = set()
antonyms = set()
for syn in synsets:
    for lemma in syn.lemmas():
        synonyms.add(lemma.name())
        if lemma.antonyms():
            antonyms.add(lemma.antonyms()[0].name())

print("Synonyms:", synonyms)
print("Antonyms:", antonyms)


''' Ankush Saha , 2341019012 , 51 '''
```

```python
# 14. (Textatistic: Readability of News Articles) Using the above techniques, download from several
# news sites current news articles on the same topic. Perform readability assessments on them to deter-
# mine which sites are the most readable. For each article, calculate the average number of words per
# sentence, the average number of characters per word and the average number of syllables per word.


import requests
from bs4 import BeautifulSoup
from textatistic import Textatistic
import re

url = "https://www.bbc.com/news/world-us-canada-68702095"
page = requests.get(url)
soup = BeautifulSoup(page.text, "html.parser")
article_text = " ".join(p.text for p in soup.find_all("p"))

analysis = Textatistic(article_text)

words = analysis.word_count
sentences = analysis.sent_count
syllables = analysis.sybl_count
characters = len(re.sub(r"\s+", "", article_text))

print("Words per sentence:", words / sentences)
print("Characters per word:", characters / words)
print("Syllables per word:", syllables / words)


''' Ankush Saha , 2341019012 , 51 '''
```
